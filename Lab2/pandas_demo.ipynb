## Groupby operations

"Group by" operations are some of the most common and powerful operations in pandas.

As discussed https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html: 

By “group by” we are referring to a process involving one or more of the following steps:
* **Splitting** the data into groups based on some criteria.
* **Applying** a function to each group independently.
* **Combining** the results into a data structure.


Out of these, the split step is the most straightforward. In fact, in many situations we may wish to split the data set into groups and do something with those groups. In the apply step, we might wish to one of the following:

* **Aggregation:** compute a summary statistic (or statistics) for each group. Some examples:
  * Compute group sums or means.
  * Compute group sizes / counts.

* **Transformation:** perform some group-specific computations and return a like-indexed object. Some examples:
  * Standardize data (zscore) within a group.
  * Filling NAs within groups with a value derived from each group.

* **Filtration:** discard some groups, according to a group-wise computation that evaluates True or False. Some examples:

  * Discard data that belongs to groups with only a few members.
  * Filter out data based on the group sum or mean.

Some combination of the above: GroupBy will examine the results of the apply step and try to return a sensibly combined result if it doesn’t fit into either of the above two categories.

We consider some simple examples.

```python
df1
```

Suppose we are interested in the two groups (here, "MIL" and "CIV")

```python
groups = df1.groupby('Type')
groups
```

``groups`` is a special object that contains all the data, separate by each group; this object is iterable

```python
for g in groups:
    print(g)
```

Most commonly, one directly applies some function to each group.

```python
df1.groupby('Type').sum()
```

Note that this operation was applied to each column for which it made sense.  In this case, ``sum()`` only works on numerical data.

```python
df1.groupby('Type').max()
```

The ``max()`` works on both numerical and string data, so the result here is a little weird.

If you only want to consider specific columns, select those.

```python
df1.groupby('Type')['Budget','SpendRate'].max()
```

## Pivot Tables

One of the most powerful features in spreadsheet programs (like Microsoft Excel) is the ability to create "pivot tables". 

A Pivot Table is used to summarise, sort, reorganise, group, count, total or average data stored in a table. It allows us to transform columns into rows and rows into columns. It allows grouping by any field (column), and using advanced calculations on them.

Pandas can do this too: https://pandas.pydata.org/pandas-docs/stable/user_guide/reshaping.html

```python
df1
```

```python
# create a pivot table of budget values, by quarter and region
df1.pivot_table('Budget',index='Quarter',columns='Office',aggfunc=sum)
```

```python
# create another pivot table, by type of construction and quarter
df1.pivot_table('Budget',index='Type',columns='Quarter',aggfunc=sum)
```

```python
# let's save that table, and then augment it
df2 = df1.pivot_table('Budget',index='Type',columns='Quarter',aggfunc=sum)
df2
```

```python
# we can also plot, provided we have enabled it in the notebook
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df2.plot(kind='bar')
```

```python
# create two new columns in the pivot table, with the min/max values respectively
df2['min']=df2.min(axis=1)
df2['max']=df2.max(axis=1)
df2
```

```python
df2.to_csv("construction_pivot.csv")
```
